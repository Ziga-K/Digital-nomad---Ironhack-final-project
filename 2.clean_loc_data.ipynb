```python
"""
Use geopy to get latitude and longitude for each city

Use latitude and longitude from dataframe to get coutry names

"""

# import geopys geolocator 
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="MyGeocoder")

# import a RateLimiter (adds a delay in query making sure not to trigger antiflood protection) and fit it with geolocator
from geopy.extra.rate_limiter import RateLimiter
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
```

```python
# get longitude and latidute with geopy
df['location'] = df['city'].progress_apply(geocode)

df['lat'] = df['location'].apply(lambda geolocator: geolocator.point[0] if geolocator else None)
df['long'] = df['location'].apply(lambda geolocator: geolocator.point[1] if geolocator else None)
```

```python
# loop trough dataframe getting lat and long and use geopy to get location details
geopy_loc = []
for i,row in df.iterrows():
    lat = row["lat"]
    long = row["long"]
    location = geolocator.reverse(f"{lat},{long}", language='en')
    geopy_loc.append(location)
```

```python
# loop to get country names 
countries = []
for i in range (len(geopy_loc)):
    countries.append((geopy_loc[i][0]).split(",")[-1])
```

```python
# save countries to dataframe column
df['country'] = countries
```

```python
import pycountry_convert as pc
```

```python
# Get continent names

continents = []#list that i'm using to build my new column
#We are getting continent codes instead of full name, i wil use this dictionary to change it
cont_codes = {'AF':"africa", 'AS':"asia", 'EU':"europe", 'NA':"north_america", 'OC':"oceania", 'SA':"south_america"}
for i,row in df.iterrows():
    try: 
        country = row["country"].strip() #some countries have spaces before or after
        country_code = pc.country_name_to_country_alpha2(country, cn_name_format="default")#we need the 2letters code for getting the continent code

        cont_code= pc.country_alpha2_to_continent_code(country_code)
        continents.append(cont_codes[f"{cont_code}"])
    except:
        continents.append("Check_continent")
```

```python
#count how many countries my code wasn't able to get the continent >>> you'll have to change it manually
continents.count("Check_continent")
```

```python
#create column from "continents" list
df["continent"] = continents
```
